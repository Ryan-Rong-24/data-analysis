```python
import pandas as pd
import numpy as np

#load files
abb = pd.read_csv('../../data_source/state-abbrevs.csv')
areas = pd.read_csv('../../data_source/state-areas.csv')
pops = pd.read_csv('../../data_source/state-population.csv')

#merge population and abbreviations
merged = pd.merge(pops,abb,how='outer',left_on='state/region',right_on='abbreviation')
print(merged)
print(merged.isnull().any())
merged = merged.drop('abbreviation',1)

# print(merged.head())
# print(abb.head())
# print(pops.head())
# print(areas.head())
```

```python
# print(np.invert(merged['population'].isnull()))
# print(merged[np.invert(merged['population'].isnull())].head())
# print(merged[merged['population'].isnull()].head())

#print state 缺失值
print(merged.loc[merged['state'].isnull(),'state/region'].unique())

#填补 state 缺失值
merged.loc[merged['state/region']=='USA','state'] = 'United States'
merged.loc[merged['state/region']=='PR','state'] = 'Puerto Rico'

#确认是否填充
print(merged.isnull().any())
```

```python
final = pd.merge(merged,areas,on='state',how='left')
#final.loc[if this column is empty ,keyword]
print(final.loc[final['area (sq. mi)'].isnull(),'state/region'].unique()) 
print(final['state'][final['area (sq. mi)'].isnull()].unique())
final.dropna(inplace=True)
for i,j in final.iterrows():
    print(j)
```

```python
data2010 = final.query("(year == 2010) and (ages == 'total')")
print(data2010) 
```

```python
data2010.set_index('state',inplace = True)
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending = False, inplace = True)
print(density.tail())
print(data2010)
```
