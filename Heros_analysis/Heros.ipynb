```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
heros = pd.read_csv('heros.csv')


heros['最大攻速'] = [float(i[:-1]) for i in heros['最大攻速']]
heros = heros.dropna(axis=1)
# heros.hist() assignment!!!@!!!!!!@@!@!@@@@!@!@!@!@!@!@!@!@!@!@!@12!@!
topthree = heros.sort_values('最高物攻',ascending=False,axis=0)
topthree.iloc[:3,:3]
heros.groupby('攻击范围').count().iloc[:,:1]
for i in heros.groupby('主要定位')['最大生命']:
    print(i)

# a = []
# for i in range(len(heros.query("(主要定位=='坦克')"))):
#     a.append(heros.query("(主要定位=='坦克')").iloc[i][1])
heros.groupby('主要定位')['最大生命'].hist(by=heros['主要定位'])
plt.title("Max Health of Tanks")
plt.xlabel('HP')
plt.ylabel('No. of tanks')
plt.show()
```
